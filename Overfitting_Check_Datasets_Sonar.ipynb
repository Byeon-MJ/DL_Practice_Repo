```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# seed 고정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 로드
df = pd.read_csv("/content/gdrive/MyDrive/Study/Everyone's DeepLearning/dataset/sonar.csv", header=None)
```

```python
df.head()
```

```python
df.info()
```

```python
dataset = df.values
X = dataset[:, 0:60]
Y_obj = dataset[:, 60]
```

```python
X = X.astype(np.float32)
```

```python
# Label Encoding
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 모델 정의
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', 
              optimizer='adam', metrics=['accuracy'])

# 모델 실행
history = model.fit(X, Y, epochs=200, batch_size=5)
print(f'Accurcy : {model.evaluate(X, Y)[1]:.4f}')
```

# 초음파 광물 예측하기 : K-Fold 교차 검증

```python
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Study/Everyone's DeepLearning/dataset/sonar.csv", header=None)
```

```python
dataset = df.values
X = dataset[:, :60].astype(np.float32)
Y_obj = dataset[:, 60]

e = LabelEncoder()
Y = e.fit_transform(Y_obj)
```

```python
# k-fold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
# 모델 생성, 컴파일, 실행
accuracy = []

for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    k_accuracy = f'{model.evaluate(X[test], Y[test])[1]:.4f}'
    accuracy.append(k_accuracy)
```

```python
print("10 fold accuracy :", accuracy)
```

```python
print('10 fold accuracy average :', np.mean([float(s) for s in accuracy]))
```
